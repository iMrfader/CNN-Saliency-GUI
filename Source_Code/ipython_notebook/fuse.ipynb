```python
img_ori = np.array(Image.open('/home/lingkun/Desktop/11_ori.jpg'))
img_add = np.array(Image.open('/home/lingkun/Desktop/new.jpg'))
imshow(img_ori)
```

```python
x,y,z = img_ori.shape
a,b,c = img_add.shape
print x,y,z,a,b,c
```

```python
img_res2 = img_ori
print img_res2.shape
# here we use pixel multiply to fuse the saliency map and original image
for i in range(a):
    for j in range(b):
        for t in range(3):
            img_res2[a-i-1][b-j-1][t] = img_ori[a-i-1][b-j-1][t]/255.0 * img_add[a-i-1][b-j-1][t]
imshow(img_res2)
```
